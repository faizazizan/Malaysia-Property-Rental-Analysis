```python
import pandas as pd

# Read the CSV file
df = pd.read_csv('rent.csv')

# Remove "RM" and "per month" from the 'monthly_rent' column
df['monthly_rent'] = df['monthly_rent'].str.replace('RM', '').str.replace(' per month', '')

# Remove spaces and commas from the 'monthly_rent' values
df['monthly_rent'] = df['monthly_rent'].str.replace(' ', '').str.replace(',', '')

# Convert 'monthly_rent' to float
df['monthly_rent'] = df['monthly_rent'].astype(float)

# Calculate the IQR (Interquartile Range)
Q1 = df['monthly_rent'].quantile(0.25)
Q3 = df['monthly_rent'].quantile(0.75)
IQR = Q3 - Q1

# Define the lower and upper bounds for outlier removal
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Remove outliers from the dataframe
df = df[(df['monthly_rent'] >= lower_bound) & (df['monthly_rent'] <= upper_bound)]

# Print the updated dataframe without outliers
print(df)
```
